```python
import pandas as pd
from time import time
```

```python
col_names = ["duration","protocol_type","service","flag","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate","label"]

kdd_data_10percent = pd.read_csv("/Users/samanthawise/Documents/Bristol/MASTERS/Data Science Toolbox/Workshops/data/kddcup.data_10_percent", header=None, names = col_names)
kdd_data_10percent.describe()
```

```python
kdd_data_10percent['label'].value_counts()
```

```python
num_features = [
    "duration","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate"
]
features = kdd_data_10percent[num_features].astype(float)
features.describe()
```

```python
from sklearn.neighbors import KNeighborsClassifier

labels = kdd_data_10percent['label'].copy()
labels[labels!='normal.'] = 'attack.'
labels.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
features[features.columns] = scaler.fit_transform(features[features.columns])
features.describe()
```

```python
clf = KNeighborsClassifier(n_neighbors = 5, algorithm = 'ball_tree', leaf_size=500)
t0 = time()
clf.fit(features,labels)
tt = time()-t0
print("Classifier trained in {} seconds".format(round(tt,3)))
```

```python
kdd_data_corrected = pd.read_csv("/Users/samanthawise/Documents/Bristol/MASTERS/Data Science Toolbox/Workshops/data/corrected", header=None, names = col_names)
kdd_data_corrected['label'].value_counts()
```

```python
kdd_data_corrected['label'][kdd_data_corrected['label']!='normal.'] = 'attack.'
kdd_data_corrected['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
kdd_data_corrected[num_features] = kdd_data_corrected[num_features].astype(float)
#kdd_data_corrected[num_features].apply(lambda x: MinMaxScaler().fit_transform(x))
kdd_data_corrected[num_features] = scaler.fit_transform(kdd_data_corrected[num_features])
kdd_data_corrected
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(
    kdd_data_corrected[num_features], 
    kdd_data_corrected['label'], 
    test_size=0.1, 
    random_state=42)
```

```python
t0 = time()
pred = clf.predict(features_test)
tt = time() - t0
print("Predicted in {} seconds".format(round(tt,3)))
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(pred, labels_test)
print("R squared is {}.".format(round(acc,4)))
```

```python
from sklearn.cluster import KMeans
k = 30
km = KMeans(n_clusters = k)


t0 = time()
km.fit(features)
tt = time()-t0
print("Clustered in {} seconds".format(round(tt,3)))
```
